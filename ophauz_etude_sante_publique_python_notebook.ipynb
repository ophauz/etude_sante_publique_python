### on prépare les dataframe sous_nutri et popu pour les merger en un nouveau dataframe  popu_plus_sous_nutrition_m1

```python
sous_nutri['Année'] = sous_nutri['Année'].replace(['2012-2014', '2013-2015', '2014-2016', '2015-2017', '2016-2018', '2017-2019'],[2013, 2014, 2015, 2016, 2017, 2018])
sous_nutri['Année']
```

```python
sous_nutri['Valeur'] = sous_nutri['Valeur'].fillna(0)
```

```python
sous_nutri['Année'] = sous_nutri['Année'].astype('int')
popu['Année'] = popu['Année'].astype('int')
```

```python
popu_plus_sous_nutrition_m1 = pd.merge(popu, sous_nutri, on = ['Zone', 'Année'], how ='left')
popu_plus_sous_nutrition_m1 = popu_plus_sous_nutrition_m1.rename(columns = {'Valeur_x':'population', 'Valeur_y': 'population_sous_nourrie'})
popu_plus_sous_nutrition_m1.head()
```

### on calcule la proportion de personnes sous-nourries en 2017

```python
proportion_sous_nutrition_2017 = round(popu_plus_sous_nutrition_m1.loc[popu_plus_sous_nutrition_m1['Année'] == 2017, 'population_sous_nourrie'].sum() * 1000 * 100 / popu_plus_sous_nutrition_m1.loc[popu_plus_sous_nutrition_m1['Année'] == 2017, 'population'].sum(), 2)
```

```python
print("En 2017, la proportion de la population mondiale en état de sous-nutrition est de " + str(proportion_sous_nutrition_2017) + " %")
```

### on calcule l'apport calorique moyen nécessaire depuis les données de la FAO : https://www.fao.org/3/y5686e/y5686e08.htm#bm08

```python
apport_moyen_femmes = (1650+3850+1750+3400+1550+3150) / 6
apport_moyen_hommes = (2100+4500+2100+4200+1700+3600) / 6
apport_moyen = round((apport_moyen_femmes + apport_moyen_hommes) / 2)
```

```python
print("L'apport calorique journalier moyen nécessaire est de " + str(apport_moyen) + " kcal par personne")
```

### on réduit et merge le dataframe dispo_ali pour créer un nouveau dataframe de travail kcal_pays_m2 

```python
kcal_pays_m2 = dispo_ali.groupby(['Zone'], as_index = False)
```

```python
kcal_pays_m2 = kcal_pays_m2['Disponibilité alimentaire (Kcal/personne/jour)'].sum()
kcal_pays_m2.head()
```

```python
popu_m2 = popu_plus_sous_nutrition_m1.drop(columns=['population_sous_nourrie'])
popu_m2.head()
```

```python
kcal_pays_popu_m2['Disponibilité alimentaire (kcal/an)'] = kcal_pays_popu_m2['Disponibilité alimentaire (Kcal/personne/jour)'] * kcal_pays_popu_m2['population'] * 365.25
kcal_pays_popu_m2.head()
```

```python
dispo_ali_globale_m2 = kcal_pays_popu_m2['Disponibilité alimentaire (kcal/an)'].sum()
dispo_ali_globale_m2
```

```python
personnes_nourrissables_m2 = round((dispo_ali_globale_m2 / (apport_moyen * 365.25)))
personnes_nourrissables_m2
```

```python
pourcentage_m2 = round(personnes_nourrissables_m2 * 100 / (popu_m2.loc[popu_m2['Année'] == 2017, 'population'].sum()))
```

```python
print("En 2017, le nombre de personnes qui auraient pu être nourries grâce à la disponibilité alimentaire globale est de \n " + str((personnes_nourrissables_m2)/1000000) + " milliards, soit " + str(pourcentage_m2) + " % de la population")
```

### on déclare les variables de calcul de proportions

```python
proportion_animale = round(dispo_ali_m4['Aliments pour animaux'].sum() * 100 / dispo_ali_m4['Disponibilité intérieure'].sum(), 2)
proportion_humaine = round(dispo_ali_m4['Nourriture'].sum() * 100 / dispo_ali_m4['Disponibilité intérieure'].sum(),2)
proportion_perdue = round(dispo_ali_m4['Pertes'].sum() * 100 / dispo_ali_m4['Disponibilité intérieure'].sum(), 2)
proportion_reste = round(100 - (proportion_animale + proportion_humaine + proportion_perdue) , 2)
```

## 4/ Toutes les infos utiles pour mettre en relief les pays qui semblent être le plus en difficulté au niveau alimentaire

```python
pays_sous_nourris_aidés = []

for i in top_pays_sous_nourris:
    for j in top_pays_aidés:
        if(i==j):
            pays_sous_nourris_aidés.append(i)
            break
            
cb_oui = len(pays_sous_nourris_aidés)

print("Sur la liste des 13 pays ayant la plus forte proportion de population en état de sous-nutrition, seul " + str(cb_oui) + ", " + str(pays_sous_nourris_aidés) + ", fait partie des pays reçevant le plus d'aide alimentaire.")
```

```python
print("Proportion de la population en état de sous-nutrition en 2017 des pays ayant reçu le plus d'aide alimentaire \ncumulée depuis 2013, par ordre décroissant : \n")

sous_nutrition_minime = []

for element in top_pays_aidés :
   
    proportion_sous_ali = proportion_sous_alimentation_2017_mel1.loc[proportion_sous_alimentation_2017_mel1['Zone'] == element].sum()
    proportion_sous_ali = proportion_sous_ali['sous-nutrition (%)']
    print(str(element) + " : " + str(round(proportion_sous_ali, 2)) + " % ")
    
    if proportion_sous_ali < 10:
        sous_nutrition_minime.append(element)
    else:
        continue
        
print("\nPays recevant les plus grosses quantités d'aide alimentaire, mais dont la proportion de la population en état de \nsous-nutrition est inférieure à 10% ou non déclarée : " + str(sous_nutrition_minime))
```

```python
pays_sous_nourris_pas_aidés = []

for element in top_pays_sous_nourris:
    if element not in top_pays_aidés:
        pays_sous_nourris_pas_aidés.append(element)

cb_non = len(pays_sous_nourris_pas_aidés)

print(str(cb_non) + " des 13 pays souffrant le plus de sous-nutrition ne reçoivent donc pas assez d'aide alimentaire. Ces pays en \nforte difficulté sont : " + str(pays_sous_nourris_pas_aidés))
```

```python
d = aide_ali.groupby(aide_ali['Pays bénéficiaire'])
d = d.sum()
d['Valeur'].describe()
```

```python
print("Quantité cumulée d'aide alimentaire reçue depuis 2013 (exprimée en tonnes) de ces pays au fort taux de sous-\nnutrition, ne faisant pas partie des pays les plus aidés : \n")

aide_minime = []
cata = []

for element in pays_sous_nourris_pas_aidés :
    aide = aide_ali.loc[aide_ali['Pays bénéficiaire'] == element].sum()
    aide = aide['Valeur']
    print(str(element) + " : " + str(aide))
    
    if aide < 3.702100e+04:
        aide_minime.append(element)
    if aide < 9.727500e+03:
        cata.append(element)
    else:
        continue
        
print("\nPays dont la proportion de la population en état de sous-nutrition est parmi les plus hautes, mais recevant des \nquantités d'aide alimentaire limitées (< 3.702100e+04 tonnes, la valeur médiane des aides versées) : " + str(aide_minime))
 
print("\nLes pays en situation critique, recevant moins de 9.727500e+03 tonnes d'aide alimentaire (la limite du premier \nquartile), sont : " + str(cata))
```
